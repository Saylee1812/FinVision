```python
import matplotlib.pyplot as plt
import seaborn as sns

#bar graph for estimated income as income band
bins=[0,100000,300000,float('inf')]
labels=['Low','Mid','High']

df['Income Band']=pd.cut(df['Estimated Income'],bins=bins,labels=labels,right=False)
#count for each band
income_counts=df['Income Band'].value_counts()
#plot graph
plt.figure(figsize=(8,6))
income_counts.plot(kind='bar',color='skyblue')
plt.title('Bar grph for Estimated Income')
plt.xlabel('Category')
plt.ylabel('No.of People')
plt.show()
        

```

```python
#distribution of values in every categorical column
from IPython.display import display
categorical_cols=df[['IAId','BRId','GenderId','Risk Weighting','Properties Owned','Amount of Credit Cards','Occupation','Fee Structure','Loyalty Classification','Nationality','Income Band']].columns

for col in categorical_cols:
    print(f"Value Counts for '{col}': (dtype: {df[col].dtype})")  # show dtype
    display(df[col].value_counts().reset_index().rename(columns={'index': col, col: 'count'}))
    print("\n")
   
```

## UNIVARIATE ANALYSIS OF CATEGORICAL COLUMNS

```python
#Distribution of categorical columns
for col in categorical_cols:
    if col == 'Occupation':
        continue
    plt.figure(figsize=(8,6))
    sns.histplot(df[col])
    plt.title(f"Distribution of values for {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
    
```

## BIVARIATE ANALYSIS OF CATEGORICAL COLUMNS

```python
for i, col in enumerate(categorical_cols):
    if col=="Occupation":
        continue
    plt.figure(i)
    sns.countplot(data=df,x=col,hue="Nationality")
```

## DISTRIBUTION OF NUMERICAL COLUMNS 

```python
numerical_cols=['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']

plt.figure(figsize=(15,10))

for i,col in enumerate(numerical_cols):
    plt.subplot(5,3,i+1)
    sns.histplot(df[col],kde=True)
    plt.title(col)
plt.tight_layout()
plt.show()

```

## HEATMAP

```python
correlation_matrix=df['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending'].corr()

plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix,annot=True,cmap='crest',fmt='.2f')
plt.title('Correlation Heatmap of Numerical Features',fontsize=16)
plt.show()
```
